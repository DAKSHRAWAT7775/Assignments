3. App routing in Flask is the mechanism to map URLs to specific functions in the Flask application. These functions are known as view functions and define what should be displayed when a user navigates to a particular URL.

We do app routing for following reasons:
(a)Define URL Patterns: Routes allow developers to define the URL structure of their web applications.
(b)Associate URLs with Functions: Routes map specific URLs to the functions that handle the requests for those URLs.
(c)Create User-Friendly URLs: Routes enable the creation of user-friendly URLs that are easy to remember and navigate.
(d)Control Application Flow: Routes help control the flow of the application, ensuring that the correct view function is called for each URL.

5. The url_for() function is used in Flask to build a URL to a specific function. This is useful for generating dynamic URLs and avoiding hardcoding URLs in the application.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return '''
    <a href="{}">Welcome Page</a>
    '''.format(url_for('welcome'))

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

if __name__ == '__main__':
    app.run()
