```python
#3. ZeroDivisionError and OverflowError are defined in the Arithmetic Error class.
try:
    result = 10 / 0 
except ZeroDivisionError as zde:
    print("Caught a ZeroDivisionError")
import math
try:
    result = math.exp(1000)  
except OverflowError as oe:
    print("Caught an OverflowError")
```

```python
'''4. 
The LookupError class in Python is a base class for exceptions that arise when a key or index is not found during a lookup operation.
LookupError itself is not meant to be instantiated directly but serves as a base class for more specific lookup-related exceptions, such as KeyError and IndexError.'''
my_dict = {'key1': 'value1', 'key2': 'value2'}
try:
    value = my_dict['key3'] 
except KeyError as ke:
    print("Caught a KeyError")
my_list = [1, 2, 3]
try:
    value = my_list[4]  
except IndexError as ie:
    print(f"Caught an IndexError")
```
